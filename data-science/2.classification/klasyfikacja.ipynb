## Klasyfikacja (SKN Data Science)

Klasyfikacja to zadanie ML, polegające na przydzielaniu etykiet różnym rzeczą na podstawie pewnych danych. Jest to jeden z problemów związanych z trenowaniem nadzorowanym.

Rodzaje klasyfikacji:
- binarna;
- wieloklasowa;
- multietykietowa;

Przykłady algorytmów klasyfikacji:
- drzewka decyzyjne;
- regresja logistyczna;
- klasyfikator random forest;
- naiwny klasyfikator bayesa;
- KNN;
- SVC;

Metryki oceny modeli:
- dokładność - procent trafnych przewidywań;
- macierz pomyłek - macierz poprawnie/błędnie zklasyfikowanych etykiet;
- precyzja - mówi jakie są przewidywania klasy pozytywnej;
- recall - ile mamy pozytywanych klasyfikacji;
- f1 - to połączenie prezycji i recall;
